# ELM forcast

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
import os.path
```

```python
file_path = '/Users/JH/Documents/GitHub/PowerForecast/ver_04/forecasting_models/ELM/forecast_result'

result_file_00 = os.path.join(file_path, 'elm_forecast_result.csv')
#result_file_01 = os.path.join(file_path, 'elm_forecast_result01.csv')
#result_file_02 = os.path.join(file_path, 'elm_forecast_result02.csv')


f_00 = pd.read_csv(result_file_00)
#f_01 = pd.read_csv(result_file_01)
#f_02 = pd.read_csv(result_file_02)


RMSE_00 = f_00['RMSE'].tolist()
#MAPE_00 = f_00['MAPE'].tolist()
#RMSE_01 = f_01['RMSE'].tolist()
#MAPE_01 = f_01['MAPE'].tolist()
#RMSE_02 = f_02['RMSE'].tolist()
#MAPE_02 = f_02['MAPE'].tolist()
```

# Total Result

```python
plot_RMSE_00 = sns.distplot(RMSE_00, norm_hist=True, kde_kws={"label":"case 00"}, color='blue')
#plot_RMSE_01 = sns.distplot(RMSE_01, norm_hist=True, kde_kws={"label":"case 01"}, color='green')
#plot_RMSE_02 = sns.distplot(RMSE_02, norm_hist=True, kde_kws={"label":"case 02"}, color='red')

plt.title('ELM', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

# Case 00

```python
plot_RMSE_00 = sns.distplot(RMSE_00, norm_hist=True, kde=True, rug=True, color='blue')
plt.title('ELM - case 00', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_00 = np.mean(RMSE_00)
std_00 = np.std(RMSE_00)

print 'number of test samples = ',
print str(len(RMSE_00))
print

print 'mean value of RMSE = ',
print str(mean_00)
print 'std value of RMSE = ',
print str(std_00)
print

print 'min RMSE = ',
print str(min(RMSE_00))
print 'max RMSE = ',
print str(max(RMSE_00))
```
