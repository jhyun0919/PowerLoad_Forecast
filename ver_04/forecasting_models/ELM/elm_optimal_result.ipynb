```python
import numpy as np
import pandas as pd
import operator

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
```

```python
result_file_00 = './elm_optimal_result00.csv'
result_file_01 = './elm_optimal_result01.csv'
result_file_02 = './elm_optimal_result02.csv'


f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)


result_00 = f_00['Acc'].tolist()
result_01 = f_01['Acc'].tolist()
result_02 = f_02['Acc'].tolist()
```

---------
