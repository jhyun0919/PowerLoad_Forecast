# ELM forcast

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
```

```python
result_file_00 = './elm_forecast_result00.csv'
result_file_01 = './elm_forecast_result01.csv'
result_file_02 = './elm_forecast_result02.csv'
result_file_03 = './elm_forecast_result03.csv'
#result_file_04 = './elm_forecast_result04.csv'

f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)
f_03 = pd.read_csv(result_file_03)
#f_04 = pd.read_csv(result_file_04)

result_00 = f_00['MAPE'].tolist()
result_01 = f_01['MAPE'].tolist()
result_02 = f_02['MAPE'].tolist()
result_03 = f_03['MAPE'].tolist()
#result_04 = f_04['MAPE'].tolist()
```

# Total Result

```python
plot_00 = sns.distplot(result_00, norm_hist=True, kde_kws={"label":"case 00"}, color='blue')
plot_01 = sns.distplot(result_01, norm_hist=True, kde_kws={"label":"case 01"}, color='green')
plot_02 = sns.distplot(result_02, norm_hist=True, kde_kws={"label":"case 02"}, color='red')
plot_03 = sns.distplot(result_03, norm_hist=True, kde_kws={"label":"case 03"}, color='purple')
#plot_04 = sns.distplot(result_04, norm_hist=True, kde_kws={"label":"case 04"}, color='orange')
plt.title('ELM', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

# Case 00

```python
plot_00 = sns.distplot(result_00, norm_hist=True, kde=True, rug=True, color='blue')
plt.title('ELM - case 00', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_00 = np.mean(result_00)
std_00 = np.std(result_00)

print 'number of test samples = ',
print str(len(result_00))
print

print 'mean value of RMSE = ',
print str(mean_00)
print 'std value of RMSE = ',
print str(std_00)
print

print 'min RMSE = ',
print str(min(result_00))
print 'max RMSE = ',
print str(max(result_00))
```
