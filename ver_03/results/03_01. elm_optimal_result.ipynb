# ELM optimal

```python
import numpy as np
import pandas as pd
import operator

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
import os.path
```

```python
file_path = '/Users/JH/Documents/GitHub/PowerForecast/ver_03/forecasting_models/ELM/optimal_data'

result_file_00 = os.path.join(file_path, 'elm_optimal_d.csv')
result_file_01 = os.path.join(file_path, 'elm_optimal_w.csv')
result_file_02 = os.path.join(file_path, 'elm_optimal_dw.csv')


f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)


RMSE_00 = f_00['RMSE'].tolist()
#MAPE_00 = f_00['MAPE'].tolist()
RMSE_01 = f_01['RMSE'].tolist()
#MAPE_01 = f_01['MAPE'].tolist()
RMSE_02 = f_02['RMSE'].tolist()
#MAPE_02 = f_02['MAPE'].tolist()
```

```python
plot_RMSE_00 = plt.plot(RMSE_00, color='blue')
plot_RMSE_01 = plt.plot(RMSE_01, color='green')
plot_RMSE_02 = plt.plot(RMSE_02, color='red')
plt.title('ELM', fontsize=13)
plt.xlabel('Hidden node number', fontsize=10)
plt.ylabel('RMSE', fontsize=10)
```

```python
plt.close()
```
