```python
def total_result(acc_type, df_00, df_01, df_02, df_03):
    idx_00 = df_00[acc_type].idxmin()-1
    idx_01 = df_01[acc_type].idxmin()-1
    idx_02 = df_02[acc_type].idxmin()-1
    idx_03 = df_03[acc_type].idxmin()-1
    
    min_00 = df_00[acc_type].iloc[idx_00]
    min_01 = df_01[acc_type].iloc[idx_01]
    min_02 = df_02[acc_type].iloc[idx_02]
    min_03 = df_03[acc_type].iloc[idx_03]
    
    print 'case_00 : idx = ' + str(idx_00) + ' / value = ' + str(min_00)
    print 'case_01 : idx = ' + str(idx_01) + ' / value = ' + str(min_01)
    print 'case_02 : idx = ' + str(idx_02) + ' / value = ' + str(min_02)
    print 'case_03 : idx = ' + str(idx_03) + ' / value = ' + str(min_03)
```

```python
def total_RMSE_graph(df_00, df_01, df_02, df_03):
    font_size = 6

    plt.subplot(221)
    plt.plot(df_00[:], color='blue', label='ELM:Case00')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)

    plt.subplot(222)
    plt.plot(df_01[30:], color='green', label='ELM:Case01')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)
    plt.ylim(0.077, 0.1)

    plt.subplot(223)
    plt.plot(df_02[30:], color='red', label='ELM:Case02')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)
    plt.ylim(0.077, 0.1)

    plt.subplot(224)
    plt.plot(df_03[30:], color='purple', label='ELM:Case03')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)
    plt.ylim(0.077, 0.1)
```

```python
def total_MAE_graph(df_00, df_01, df_02, df_03):
    font_size = 6

    plt.subplot(221)
    plt.plot(df_00[:], color='blue', label='ELM:Case00')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)

    plt.subplot(222)
    plt.plot(df_01[30:], color='green', label='ELM:Case01')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)
    plt.ylim(10000, 13000)

    plt.subplot(223)
    plt.plot(df_02[30:], color='red', label='ELM:Case02')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)
    plt.ylim(10000, 13000)

    plt.subplot(224)
    plt.plot(df_03[30:], color='purple', label='ELM:Case03')
    plt.tick_params(labelsize=font_size)
    plt.legend(loc=1)
    plt.ylim(10000, 13000)
```
