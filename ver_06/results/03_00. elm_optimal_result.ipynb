```python
def total_result(acc_type, df_00, df_01, df_02, df_03, df_04, df_05, df_06, df_07, df_08):
    idx_00 = df_00[acc_type].idxmin()-1
    idx_01 = df_01[acc_type].idxmin()-1
    idx_02 = df_02[acc_type].idxmin()-1
    idx_03 = df_03[acc_type].idxmin()-1
    idx_04 = df_04[acc_type].idxmin()-1
    idx_05 = df_05[acc_type].idxmin()-1
    idx_06 = df_06[acc_type].idxmin()-1
    idx_07 = df_07[acc_type].idxmin()-1
    idx_08 = df_08[acc_type].idxmin()-1
    
    min_00 = df_00[acc_type].iloc[idx_00]
    min_01 = df_01[acc_type].iloc[idx_01]
    min_02 = df_02[acc_type].iloc[idx_02]
    min_03 = df_03[acc_type].iloc[idx_03]
    min_04 = df_04[acc_type].iloc[idx_04]
    min_05 = df_05[acc_type].iloc[idx_05]
    min_06 = df_06[acc_type].iloc[idx_06]
    min_07 = df_07[acc_type].iloc[idx_07]
    min_08 = df_08[acc_type].iloc[idx_08]
    
    print 'case_00 : idx = ' + str(idx_00) + ' / value = ' + str(min_00)
    print 'case_01 : idx = ' + str(idx_01) + ' / value = ' + str(min_01)
    print 'case_02 : idx = ' + str(idx_02) + ' / value = ' + str(min_02)
    print 'case_03 : idx = ' + str(idx_03) + ' / value = ' + str(min_03)
    print 'case_04 : idx = ' + str(idx_04) + ' / value = ' + str(min_04)
    print 'case_05 : idx = ' + str(idx_05) + ' / value = ' + str(min_05)
    print 'case_06 : idx = ' + str(idx_06) + ' / value = ' + str(min_06)
    print 'case_07 : idx = ' + str(idx_07) + ' / value = ' + str(min_07)
    print 'case_08 : idx = ' + str(idx_08) + ' / value = ' + str(min_08)
```
