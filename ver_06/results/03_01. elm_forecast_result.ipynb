```python
def total_result(acc_type, data_type, list_00, list_01, list_02, list_03, list_04, list_05, list_06, list_07, list_08):
    train_mean_00 = np.mean(list_00)
    train_std_00 = np.std(list_00)
    
    train_mean_01 = np.mean(list_01)
    train_std_01 = np.std(list_01)
    
    train_mean_02 = np.mean(list_02)
    train_std_02 = np.std(list_02)
    
    train_mean_03 = np.mean(list_03)
    train_std_03 = np.std(list_03)
    
    train_mean_04 = np.mean(list_04)
    train_std_04 = np.std(list_04)
    
    train_mean_05 = np.mean(list_05)
    train_std_05 = np.std(list_05)
    
    train_mean_06 = np.mean(list_06)
    train_std_06 = np.std(list_06)
    
    train_mean_07 = np.mean(list_07)
    train_std_07 = np.std(list_07)
    
    train_mean_08 = np.mean(list_08)
    train_std_08 = np.std(list_08)

    
    print data_type
    
    print 'Case 00 [' + acc_type + '/std] = ',
    print str(train_mean_00) + ' / ' + str(train_std_00)
    
    print 'Case 01 [' + acc_type + '/std] = ',
    print str(train_mean_01) + ' / ' + str(train_std_01)
    
    print 'Case 02 [' + acc_type + '/std] = ',
    print str(train_mean_02) + ' / ' + str(train_std_02)
    
    print 'Case 03 [' + acc_type + '/std] = ',
    print str(train_mean_03) + ' / ' + str(train_std_03)
    
    print 'Case 04 [' + acc_type + '/std] = ',
    print str(train_mean_04) + ' / ' + str(train_std_04)
    
    print 'Case 05 [' + acc_type + '/std] = ',
    print str(train_mean_05) + ' / ' + str(train_std_05)
    
    print 'Case 06 [' + acc_type + '/std] = ',
    print str(train_mean_06) + ' / ' + str(train_std_06)
    
    print 'Case 07 [' + acc_type + '/std] = ',
    print str(train_mean_07) + ' / ' + str(train_std_07)
    
    print 'Case 08 [' + acc_type + '/std] = ',
    print str(train_mean_08) + ' / ' + str(train_std_08)
```

```python
def total_RMSE_graph(acc_type, data_type, list_00, list_01, list_02, list_03, list_04, list_05, list_06, list_07, list_08):
    if data_type is 'Train':
        plot_train_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"})
        plot_train_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"})
        plot_train_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"})
        plot_train_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"})
        plot_train_04 = sns.distplot(list_04, norm_hist=True, kde_kws={"label":"case 04"})
        plot_train_05 = sns.distplot(list_05, norm_hist=True, kde_kws={"label":"case 05"})
        plot_train_06 = sns.distplot(list_06, norm_hist=True, kde_kws={"label":"case 06"})
        plot_train_07 = sns.distplot(list_07, norm_hist=True, kde_kws={"label":"case 07"})
        plot_train_08 = sns.distplot(list_08, norm_hist=True, kde_kws={"label":"case 08"})
    elif data_type is 'Test':
        plot_test_00 = sns.distplot(list_00, norm_hist=True, kde_kws={"label":"case 00"})
        plot_test_01 = sns.distplot(list_01, norm_hist=True, kde_kws={"label":"case 01"})
        plot_test_02 = sns.distplot(list_02, norm_hist=True, kde_kws={"label":"case 02"})
        plot_test_03 = sns.distplot(list_03, norm_hist=True, kde_kws={"label":"case 03"})
        plot_test_04 = sns.distplot(list_04, norm_hist=True, kde_kws={"label":"case 04"})
        plot_test_05 = sns.distplot(list_05, norm_hist=True, kde_kws={"label":"case 05"})
        plot_test_06 = sns.distplot(list_06, norm_hist=True, kde_kws={"label":"case 06"})
        plot_test_07 = sns.distplot(list_07, norm_hist=True, kde_kws={"label":"case 07"})
        plot_test_08 = sns.distplot(list_08, norm_hist=True, kde_kws={"label":"case 08"})
    
    plt.title('ELM', fontsize=13)
    plt.xlabel(acc_type + ' ' + data_type, fontsize=10)
    plt.ylabel('Density', fontsize=10)

    #plt.xlim(0.06, 0.17)
    #plt.ylim(0, 5000)
```
