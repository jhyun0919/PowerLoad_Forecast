# H-ELM optimal

```python
import numpy as np
import pandas as pd
import operator

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
import os.path
import sys
sys.path.insert(0, '/Users/JH/Documents/GitHub/PowerForecast/ver_05/util')
import app as ap
```

```python
optimal_file_path = '/Users/JH/Documents/GitHub/PowerForecast/ver_05/forecasting_models/HELM/optimal_data'
area = 'ACTEWAGL'
layer = 'layer1'
file_path = os.path.join(optimal_file_path, area)
file_path = os.path.join(file_path, layer)

result_file_00 = os.path.join(file_path, 'helm_optimal_wd.csv')
result_file_01 = os.path.join(file_path, 'helm_optimal_lwd.csv')
result_file_02 = os.path.join(file_path, 'helm_optimal_lww.csv')
result_file_03 = os.path.join(file_path, 'helm_optimal_lwdw.csv')


f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)
f_03 = pd.read_csv(result_file_03)


RMSE_00 = f_00['RMSE'].tolist()
RMSE_01 = f_01['RMSE'].tolist()
RMSE_02 = f_02['RMSE'].tolist()
RMSE_03 = f_03['RMSE'].tolist()
```

```python
df_00 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_00)))
df_01 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_01)))
df_02 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_02)))
df_03 = ap.Numpy2Pandas.add_number_index(ap.Numpy2Pandas.acc_data2df(np.array(RMSE_03)))
```

```python
font_size = 9

plt.subplot(221)
plt.plot(df_00[:], color='blue')
plt.title('case 00', fontsize=font_size)

plt.subplot(222)
plt.plot(df_01[30:], color='green')
plt.ylim(0.077, 0.099)
plt.title('case 01', fontsize=font_size)

plt.subplot(223)
plt.plot(df_02[30:], color='red')
plt.ylim(0.077, 0.099)
plt.title('case 02', fontsize=font_size)

plt.subplot(224)
plt.plot(df_03[30:], color='purple')
plt.ylim(0.077, 0.099)
plt.title('case 03', fontsize=font_size)
```

```python
plt.close()
```
