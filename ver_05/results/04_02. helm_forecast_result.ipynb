# H-ELM

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
import os.path
```

```python
forecast_result_file_path = '/Users/JH/Documents/GitHub/PowerForecast/ver_05/forecasting_models/HELM/forecast_result'
area = 'ACTEWAGL'
layer = 'layer2'
file_path = os.path.join(forecast_result_file_path, area)
file_path = os.path.join(file_path, layer)

result_file_00 = os.path.join(file_path, 'helm_forecast_result_wd.csv')
result_file_01 = os.path.join(file_path, 'helm_forecast_result_lwd.csv')
result_file_02 = os.path.join(file_path, 'helm_forecast_result_lwdw.csv')
result_file_03 = os.path.join(file_path, 'helm_forecast_result_lwdwp.csv')


f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)
f_03 = pd.read_csv(result_file_03)



result_00 = f_00['RMSE'].tolist()
result_01 = f_01['RMSE'].tolist()
result_02 = f_02['RMSE'].tolist()
result_03 = f_03['RMSE'].tolist()
```

# Total Result

```python
plot_00 = sns.distplot(result_00, norm_hist=True, kde_kws={"label":"case 00"}, color='blue')
plot_01 = sns.distplot(result_01, norm_hist=True, kde_kws={"label":"case 01"}, color='green')
plot_02 = sns.distplot(result_02, norm_hist=True, kde_kws={"label":"case 02"}, color='red')
plot_03 = sns.distplot(result_03, norm_hist=True, kde_kws={"label":"case 02"}, color='purple')

plt.title('H-ELM', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

# Case 00

```python
plot_00 = sns.distplot(result_00, norm_hist=True, kde=True, rug=True, color='blue')
plt.title('H-ELM - case 00', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
mean_00 = np.mean(result_00)
std_00 = np.std(result_00)

print 'number of test samples = ',
print str(len(result_00))
print

print 'mean value of RMSE = ',
print str(mean_00)
print 'std value of RMSE = ',
print str(std_00)
print

print 'min RMSE = ',
print str(min(result_00))
print 'max RMSE = ',
print str(max(result_00))
```
