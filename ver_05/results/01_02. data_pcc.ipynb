# Pearson's Correlation Coefficient

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns

import sys
sys.path.insert(0, '/Users/JH/Documents/GitHub/PowerForecast/ver_05/util')
import app as ap
import pcc as pcc
```

```python
load_data = ap.ReadCSV.load()
area_names = sorted(load_data.iterkeys())
selected_list = [0, 2, 3, 4, 7]
for i in xrange(0, len(selected_list)):
    print area_names[selected_list[i]]
```

```python
def plot_pcc_score(area_idx, list_len):
    print area_names[area_idx]
    
    for plot_idx in xrange(0, 2):
        plt.subplot(2, 1, (plot_idx + 1))
        x = np.arange(0, list_len)
        if plot_idx == 0:
            pcc_list = pcc.get_list(load_data[area_names[area_idx]], pcc.day_type, list_len)
            plt.plot(x, pcc_list, '-ro')
            plt.xlabel('Days', fontsize=10)
            plt.title(area_names[area_idx] + ' - PCC', fontsize=13)
        if plot_idx == 1:
            pcc_list = pcc.get_list(load_data[area_names[area_idx]], pcc.week_type, list_len)
            plt.plot(x, pcc_list, '-o')
            plt.xlabel('Weeks', fontsize=10)
        plt.ylabel('PCC', fontsize=10)
    
```
