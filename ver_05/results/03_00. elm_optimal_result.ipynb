# ELM optimal

```python
import numpy as np
import pandas as pd
import operator

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
import os.path
```

```python
optimal_file_path = '/Users/JH/Documents/GitHub/PowerForecast/ver_05/forecasting_models/ELM/optimal_data'
area = 'ACTEWAGL'
file_path = os.path.join(optimal_file_path, area)

result_file_00 = os.path.join(file_path, 'elm_optimal_wd.csv')
result_file_01 = os.path.join(file_path, 'elm_optimal_lwd.csv')
result_file_02 = os.path.join(file_path, 'elm_optimal_lwdw.csv')
result_file_03 = os.path.join(file_path, 'elm_optimal_lwdwp.csv')


f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)
f_03 = pd.read_csv(result_file_03)


RMSE_00 = f_00['RMSE'].tolist()
RMSE_01 = f_01['RMSE'].tolist()
RMSE_02 = f_02['RMSE'].tolist()
RMSE_03 = f_03['RMSE'].tolist()
```

```python
plot_RMSE_00 = plt.plot(RMSE_00, color='blue')
plot_RMSE_01 = plt.plot(RMSE_01, color='green')
plot_RMSE_02 = plt.plot(RMSE_02, color='red')
plot_RMSE_03 = plt.plot(RMSE_02, color='purple')

plt.title('ELM', fontsize=13)
plt.xlabel('Hidden node number', fontsize=10)
plt.ylabel('RMSE', fontsize=10)
```

```python
plt.close()
```
