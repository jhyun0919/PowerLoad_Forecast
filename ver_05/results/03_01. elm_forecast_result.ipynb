# ELM forcast

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
import os.path
```

```python
forecast_result_file_path = '/Users/JH/Documents/GitHub/PowerForecast/ver_05/forecasting_models/ELM/forecast_result'
area = 'ACTEWAGL'
file_path = os.path.join(forecast_result_file_path, area)

result_file_00 = os.path.join(file_path, 'elm_forecast_result_wd.csv')
result_file_01 = os.path.join(file_path, 'elm_forecast_result_lwd.csv')
result_file_02 = os.path.join(file_path, 'elm_forecast_result_lww.csv')
result_file_03 = os.path.join(file_path, 'elm_forecast_result_lwdw.csv')



f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)
f_03 = pd.read_csv(result_file_03)


Training_RMSE_00 = f_00['Training'].tolist()
Testing_RMSE_00 = f_00['Testing'].tolist()
Training_RMSE_01 = f_01['Training'].tolist()
Testing_RMSE_01 = f_01['Testing'].tolist()
Training_RMSE_02 = f_02['Training'].tolist()
Testing_RMSE_02 = f_02['Testing'].tolist()
Training_RMSE_03 = f_03['Training'].tolist()
Testing_RMSE_03 = f_03['Testing'].tolist()
```

# Total Result

### Training RMSE summary

```python
plot_train_00 = sns.distplot(Training_RMSE_00, norm_hist=True, kde_kws={"label":"case 00:train"}, color='#658cbb')
plot_train_01 = sns.distplot(Training_RMSE_01, norm_hist=True, kde_kws={"label":"case 01:train"}, color='#7bb274')
plot_train_02 = sns.distplot(Training_RMSE_02, norm_hist=True, kde_kws={"label":"case 02:train"}, color='#d3494e')
plot_train_03 = sns.distplot(Training_RMSE_03, norm_hist=True, kde_kws={"label":"case 03:train"}, color='#916e99')

plt.title('ELM', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
train_mean_00 = np.mean(Training_RMSE_00)
train_std_00 = np.std(Training_RMSE_00)
train_mean_01 = np.mean(Training_RMSE_01)
train_std_01 = np.std(Training_RMSE_01)
train_mean_02 = np.mean(Training_RMSE_02)
train_std_02 = np.std(Training_RMSE_02)
train_mean_03 = np.mean(Training_RMSE_03)
train_std_03 = np.std(Training_RMSE_03)

print 'Case 00 [RMSE/std] = ',
print str(train_mean_00) + ' / ' + str(train_std_00)
print 'Case 01 [RMSE/std] = ',
print str(train_mean_01) + ' / ' + str(train_std_01)
print 'Case 02 [RMSE/std] = ',
print str(train_mean_02) + ' / ' + str(train_std_02)
print 'Case 03 [RMSE/std] = ',
print str(train_mean_03) + ' / ' + str(train_std_03)
```

### Testing RMSE summary

```python
plot_test_00 = sns.distplot(Testing_RMSE_00, norm_hist=True, kde_kws={"label":"case 00:test"}, color='blue')
plot_test_01 = sns.distplot(Testing_RMSE_01, norm_hist=True, kde_kws={"label":"case 01:test"}, color='green')
plot_test_02 = sns.distplot(Testing_RMSE_02, norm_hist=True, kde_kws={"label":"case 02:test"}, color='red')
plot_test_03 = sns.distplot(Testing_RMSE_03, norm_hist=True, kde_kws={"label":"case 03:test"}, color='purple')

plt.title('ELM', fontsize=13)
plt.xlabel('RMSE score', fontsize=10)
plt.ylabel('Density', fontsize=10)
```

```python
plt.close()
```

```python
test_mean_00 = np.mean(Testing_RMSE_00)
test_std_00 = np.std(Testing_RMSE_00)
test_mean_01 = np.mean(Testing_RMSE_01)
test_std_01 = np.std(Testing_RMSE_01)
test_mean_02 = np.mean(Testing_RMSE_02)
test_std_02 = np.std(Testing_RMSE_02)
test_mean_03 = np.mean(Testing_RMSE_03)
test_std_03 = np.std(Testing_RMSE_03)

print 'Case 00 [RMSE/std] = ',
print str(test_mean_00) + ' / ' + str(test_std_00)
print 'Case 01 [RMSE/std] = ',
print str(test_mean_01) + ' / ' + str(test_std_01)
print 'Case 02 [RMSE/std] = ',
print str(test_mean_02) + ' / ' + str(test_std_02)
print 'Case 03 [RMSE/std] = ',
print str(test_mean_03) + ' / ' + str(test_std_03)
```
