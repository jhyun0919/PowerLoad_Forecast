# ELM optimal

```python
import numpy as np
import pandas as pd
import operator

import matplotlib.pyplot as plt
%matplotlib notebook

import seaborn as sns
```

```python
result_file_00 = './elm_optimal_result00.csv'
result_file_01 = './elm_optimal_result01.csv'
result_file_02 = './elm_optimal_result02.csv'
result_file_03 = './elm_optimal_result03.csv'
result_file_04 = './elm_optimal_result04.csv'

f_00 = pd.read_csv(result_file_00)
f_01 = pd.read_csv(result_file_01)
f_02 = pd.read_csv(result_file_02)
f_03 = pd.read_csv(result_file_03)
f_04 = pd.read_csv(result_file_04)

result_00 = f_00['MAPE'].tolist()
result_01 = f_01['MAPE'].tolist()
result_02 = f_02['MAPE'].tolist()
result_03 = f_03['MAPE'].tolist()
result_04 = f_04['MAPE'].tolist()
```

# Case 00 optimal range

```python
plot_00 = plt.plot(result_00, color='blue')
plt.title('ELM', fontsize=13)
plt.xlabel('Hidden node number', fontsize=10)
plt.ylabel('RMSE', fontsize=10)
```

```python
plt.close()
```

```python
min_index, min_value = min(enumerate(result_00), key=operator.itemgetter(1))
print 'optimal hidden node number = ',
print min_index
```
